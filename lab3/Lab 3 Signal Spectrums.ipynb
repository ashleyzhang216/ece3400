```python
import numpy as np
from matplotlib import pyplot as plt

data = []

with open('500.txt') as my_file:
    for line in my_file:
        data.append(line)
        
data = [int(x) for x in data]

x = np.arange(1,len(data)+1,1)

plt.figure(figsize=(12,8))
plt.xlabel('bin #', fontsize=16)
plt.ylabel('Amplitude (a.u.)', fontsize=16)
plt.title('Nano Spectrum for 500 Hz Signal', fontsize=20)
plt.stem(x, data, use_line_collection=True)
```
